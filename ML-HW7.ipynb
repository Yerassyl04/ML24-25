```python
import numpy as np

def f(x):
    return (x - 3)**2 + 5

def gradient_f(x):
    return 2 * (x - 3)

x0 = 0  
n = 0.1 

# Хранение значений x и f(x)
x_values = [x0]
f_values = [f(x0)]


x = x0
for a in range(5):
    x = x - n * gradient_f(x)  
    x_values.append(x)
    f_values.append(f(x))

for i in range(len(x_values)):
    print(f"{i}: x = {x_values[i]:.4f}, f(x) = {f_values[i]:.4f}")
```

```python
import numpy as np

def f(x, y):
    return (x + y)**2 + (y - 3)**2

def gradient_f(x, y):
    dx = 2 * (x + y) 
    dy = 2 * (x + y) + 2 * (y - 3)
    return np.array([dx, dy])


x, y = 0, 0
n = 0.2

for i in range(3):
    grad = gradient_f(x, y) 
    x -= n * grad[0]  
    y -= n * grad[1] 
    print(f"x = {x:.3f}, y = {y:.3f}, gradient_f(x, y) = {f(x, y):.3f}")

```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**4 - 3*x**3 + 2

def gradient_f(x):
    return 4*x**3 - 9*x**2 

x = 6  
n = 0.01  

x_values = []
f_values = []


for i in range(100): # Градиентный спуск
    x_values.append(x)
    f_values.append(f(x))
    
    gradient = gradient_f(x)  
    x = x - n * gradient  

    print(f"Итерация {i+1}: x = {x:.4f}, f(x) = {f(x):.4f}")

```
