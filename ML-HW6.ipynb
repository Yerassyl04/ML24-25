# Задача 1: Математика нейронных сетей


Условие: Рассмотрим простую нейронную сеть с одним входом, одним выходом и одним скрытым слоем, состоящим из двух нейронов. Придумайте свою сеть и напишите математические выражения (как на 28м слайде).

```python
import numpy as np

# Функция активации 
def sigmoid(x):
    return 1 / (1 + np.exp(-x))


inputs = np.array([1, 0])  # Входные значения

weights_input_hidden = np.array([[0.45, -0.12], [0.78, 0.13]]) 
weights_hidden_output = np.array([1.5, -2.3])  


H1input = inputs[0] * weights_input_hidden[0][0] + inputs[1] * weights_input_hidden[0][1]
H2input = inputs[0] * weights_input_hidden[1][0] + inputs[1] * weights_input_hidden[1][1]
H1output = sigmoid(H1input)
H2output = sigmoid(H2input)


O1input = H1output * weights_hidden_output[0] + H2output * weights_hidden_output[1]
O1output = sigmoid(O1input)


print(f'H1input = {H1input:.2f}, H1output = {H1output:.2f}')
print(f'H2input = {H2input:.2f}, H2output = {H2output:.2f}')
print(f'O1input = {O1input:.2f}, O1output = {O1output:.2f}')
```

Два нейрона которые получают входные данные и применяют к ним свои весовые коэффициенты и смещения.

# Задача 3: Основы PyTorch


Условие: Напишите простой код на PyTorch для создания однослойной нейронной сети, которая принимает на вход один параметр и выдает один выход.
Создайте класс для модели с одним линейным слоем.
Инициализируйте модель, определите функцию потерь (например, MSE) и оптимизатор (например, SGD).
Напишите цикл обучения на 10 эпох, используя случайные данные.


```python
import torch
import torch.nn as nn
import torch.optim as optim


class Model(nn.Module):
    def __init__(self):
        super(Model, self).__init__()
        self.linear = nn.Linear(1, 1)  # Простая модель с одним линейным слоем

    def forward(self, x):
        return self.linear(x)

model = Model()
criterion = nn.MSELoss()  # Среднеквадратичная ошибка
optimizer = optim.SGD(model.parameters(), lr=0.01)  # Стохастический градиентный спуск


inputs = torch.randn(100, 1)
targets = torch.randn(100, 1)


for epoch in range(10):
    outputs = model(inputs)  # Прямое распространение
    loss = criterion(outputs, targets)  # Вычисление ошибки

    optimizer.zero_grad()  # Обнуление градиентов
    loss.backward()  # Обратное распространение
    optimizer.step()  # Шаг оптимизации

    print(f'Loss: {loss.item():.4f}')  # Вывод ошибки на каждой эпохе
```
