Факториал: Напишите программу, которая запрашивает у пользователя число и вычисляет его факториал.

```python
def fact(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * fact(n - 1)

number = int(input("enter: "))
print(fact(number))
```

Фибоначчи: Напишите программу, которая генерирует первые N чисел последовательности Фибоначчи, где N запрашивается у пользователя.


```python
n = int(input("n: "))

def fibonacci(n):
    f = []
    a, b = 0, 1
    for _ in range(n):
        f.append(a)  
        a, b = b, a + b  

    return f


fib = fibonacci(n)
print(fib)
```

Поиск простых чисел: Напишите программу, которая находит и выводит все простые числа в заданном диапазоне.


```python
rows = int(input("rows: "))
cols = int(input("cols: "))


def input_matrix(rows, cols):
    matrix = []
    for i in range(rows):
        row = list(map(int, input(f"enter row: ").split()))
        matrix.append(row)
    return matrix

print("1:")
matrix1 = input_matrix(rows, cols)
print("2:")
matrix2 = input_matrix(rows, cols)

result = [[matrix1[i][j] + matrix2[i][j] for j in range(cols)] for i in range(rows)]

for row in result:
    print(row)
```

Конвертер температур: Напишите программу, которая запрашивает у пользователя температуру в Цельсиях и конвертирует ее в Фаренгейты, а также наоборот. Выведите результат на экран.

```python
def ctof(celsius):
    return (celsius * 9/5) + 32

def ftoc(fahrenheit):
    return (fahrenheit - 32) * 5/9

choice = int(input("1) cel to far or 2) far to cels: "))

if choice == 1:
    celsius = float(input(""))
    fahrenheit = ctof(celsius)
    print(fahrenheit)
elif choice == 2:
    fahrenheit = float(input(""))
    celsius = ftoc(fahrenheit)
    print(celsius)
else:
    print("eror")
```
