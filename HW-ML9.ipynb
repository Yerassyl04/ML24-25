Фильтрация данных по критерию
У вас есть датафрейм с информацией о студентах: их имя, возраст и оценки по предметам (например, "Математика", "Информатика"). Отфильтруйте студентов, у которых оценка по математике выше 80 баллов.


```python
import pandas as pd

d1 = {
    'Имя': ['Ер', 'Асыл', 'Нур', 'Даулет'],
    'Возраст': [20, 20, 19, 21],
    'Матем': [85, 79, 90, 60],
    'Информ': [95, 88, 68, 92]
}
df1 = pd.DataFrame(d1)

filt = df1[df1['Матем'] > 80]
print(filt)
```

Извлечение и работа с датами
У вас есть датафрейм с данными о заказах интернет-магазина. В нем есть столбец с датой заказа в формате YYYY-MM-DD. Добавьте новый столбец, в котором укажите, в каком месяце был сделан заказ.


```python
d2= {
    'Заказ': [1, 2, 3, 4],
    'Дата': ['2022-01-10', '2023-01-10', '2023-02-10', '2023-03-10']
}
df2 = pd.DataFrame(d2)

df2['Дата'] = pd.to_datetime(df2['Дата'])

df2['Месяц'] = df2['Дата'].dt.month
print(df2)
```

Группировка и подсчет средних значений
В датафрейме содержится информация о сотрудниках компании: отдел, должность и зарплата. Найдите среднюю зарплату для каждого отдела.


```python
import pandas as pd

d3 = {
    'Отдел': ['Финансы', 'Маркетинг', 'Логистика', 'Юридический', 'Финансы'],
    'Должность': ['Аналитик', 'PR-специалист', 'Менеджер по поставкам', 'Юрист', 'Бухгалтер'],
    'Зарплата': [130000, 95000, 120000, 115000, 105000]
}
df3 = pd.DataFrame(d3)

# Группировка по отделу и расчет средней зарплаты
avgs = df3.groupby('Отдел')['Зарплата'].mean()
avgs
```

Объединение датафреймов
У вас есть два датафрейма с информацией о студентах. Один содержит их имя и возраст, а второй — имя и оценки по математике. Объедините два датафрейма по имени, чтобы получить полную информацию о студентах.


```python
d4 = pd.DataFrame({
    'Имя': ['Ер', 'Асыл', 'Нур', 'Даулет'],
    'Возраст': [20, 20, 19, 21]
})    
d5 = pd.DataFrame({
    'Имя': ['Ер', 'Асыл', 'Нур', 'Даулет'],
    'Матем': [85, 79, 90, 60]
})

full = pd.merge(d4, d5, on='Имя')
print(full)
```

Отправка GET-запроса и работа с JSON
Отправьте GET-запрос к API сайта https://jsonplaceholder.typicode.com/posts. Выведите первые пять записей из возвращенного JSON-ответа.


```python
import requests

zapr = requests.get("https://jsonplaceholder.typicode.com/posts")
data = zapr.json()

for post in data[:5]:
    print(post)
```

Поиск информации по параметрам
Используя API https://jsonplaceholder.typicode.com/users, найдите пользователя с именем "Leanne Graham". Выведите его email и адрес (street, suite, city, zipcode).


```python
zapr2 = requests.get("https://jsonplaceholder.typicode.com/users")
users = zapr2.json()

for user in users:
    if user["name"] == "Leanne Graham":
        email = user["email"]
        address = user["address"]
        print("mail:", email, "address:", address["street"], address["suite"], address["city"], address["zipcode"])
```

Обработка пропущенных значений
У вас есть датафрейм с данными о продажах, где некоторые значения в колонках "Цена" и "Количество" отсутствуют. Замените пропущенные значения в колонке "Цена" на среднюю цену, а пропущенные значения в колонке "Количество" — на медиану количества

```python
import pandas as pd

d6 = {
    'Цена': [100, 200, None, 150, None, 125],
    'Количество': [10, None, 5, None, 7, 3]
}

df6 = pd.DataFrame(d6)
df6['Цена'].fillna(df6['Цена'].mean(), inplace=True)
df6['Количество'].fillna(df6['Количество'].median(), inplace=True)

print(df6)
```
