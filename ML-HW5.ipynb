# 2. Задача: Магазин продуктов

Уровень: Средний
Создайте класс Product, который будет представлять продукт в магазине. У него должны быть атрибуты:
name (название),
price (цена),
quantity (количество).
Добавьте метод get_total_price(), который будет возвращать общую стоимость данного продукта (цена умноженная на количество).
Затем создайте класс ShoppingCart, который будет представлять корзину для покупок. В нем должны быть методы:
add_product(product) — добавляет продукт в корзину,
remove_product(product) — удаляет продукт из корзины,
get_total_cart_price() — возвращает общую стоимость всех продуктов в корзине.


```python
class Product:
    def __init__(self, name, price, quantity):
        self.name = name
        self.price = price
        self.quantity = quantity

    def get_total_price(self):
        return self.price * self.quantity

class ShoppingCart:
    def __init__(self):
        self.products = []

    def add_product(self, product):
        self.products.append(product)

    def remove_product(self, product):
        self.products.remove(product)

    def get_total_cart_price(self):
        return sum([product.get_total_price() for product in self.products])
```

# 3. Задача: Банковский аккаунт

Уровень: Средний
Создайте класс BankAccount, который будет представлять банковский аккаунт. У него должны быть атрибуты:
owner (владелец),
balance (баланс, изначально 0).
Добавьте методы:
deposit(amount) — пополнение баланса на указанную сумму,
withdraw(amount) — снятие денег, при этом нельзя снимать больше, чем есть на балансе,
get_balance() — возвращает текущий баланс.


```python
class BankAccount:
    def __init__(self, owner):
        self.owner = owner
        self.balance = 0

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
        else:
            print("Сумма депозита должна быть положительной!")

    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
        else:
            print("Недостаточно средств на счете!")

    def get_balance(self):
        return self.balance
```

2 Пополните баланс, снимите деньги и выведите текущий баланс.


```python
client1=BankAccount("Yerassyl")
```

```python
client1.deposit(10000)
```

```python
client1.withdraw(4500)
```

```python
print(client1.get_balance())
```

# 4. Задача: Библиотека книг


Уровень: Продвинутый
Создайте класс Book, который будет представлять книгу. У книги должны быть атрибуты:
title (название),
author (автор),
year (год выпуска).
Создайте класс Library, который будет хранить книги. У библиотеки должны быть методы:
add_book(book) — добавляет книгу в библиотеку,
remove_book(book) — удаляет книгу из библиотеки,
find_books_by_author(author) — находит все книги данного автора.


```python
class Book:
    def __init__(self, title, author, year):
        self.title = title
        self.author = author
        self.year = year

class Library:
    def __init__(self):
        self.books = []

    def add_book(self, book):
        self.books.append(book)

    def remove_book(self, book):
        self.books.remove(book)

    def find_books_by_author(self, author):
        return [book for book in self.books if book.author == author]
```

1 Добавьте несколько книг в библиотеку.
2 Найдите все книги определенного автора.


```python
book1 = Book("Қылкөпір", "Қабдеш Жұмаділов", 2005)
book2 = Book("Қаржыгер", "Теодор Драйзер",1930)
book3 = Book("Тағдыр", "Қабдеш Жұмаділов", 1999)
book4 = Book("Абай жолы 1", "Мұхтар Әуезов",1950)
book5 = Book("Абай жолы 2", "Мұхтар Әуезов",1952)

library = Library()
library.add_book(book1)
library.add_book(book2)
library.add_book(book3)
library.add_book(book4)
library.add_book(book5)
```

```python
zakaz1 = library.find_books_by_author("Қабдеш Жұмаділов")
```

# 5. Задача: Игра "Камень, ножницы, бумага"


Уровень: Продвинутый
Создайте класс Player, который будет представлять игрока в игру "Камень, ножницы, бумага". У игрока должны быть атрибуты:
name (имя),
choice (выбор игрока: "камень", "ножницы" или "бумага").
Создайте класс Game, который будет проводить игру между двумя игроками. У него должен быть метод:
play(player1, player2) — который выводит победителя на основе их выбора.
python


```python
class Player:
    def __init__(self, name):
        self.name = name
        self.choice = None

    def make_choice(self, choice):
        if choice in ["камень", "ножницы", "бумага"]:
            self.choice = choice
        else:
            print("Неверный выбор!")

class Game:
    def play(self, player1, player2):
        outcomes = {
            ("камень", "ножницы"): player1,
            ("ножницы", "бумага"): player1,
            ("бумага", "камень"): player1,
            ("ножницы", "камень"): player2,
            ("бумага", "ножницы"): player2,
            ("камень", "бумага"): player2,
        }
        
        if player1.choice == player2.choice:
            print("Ничья!")
        else:
            winner = outcomes.get((player1.choice, player2.choice))
            print(f"Победитель: {winner.name}")
```

1 Создайте двух игроков.
2 Проведите игру и определите победителя.


```python
player1 = Player("Мақаш")
player2 = Player("Сейдалы")


player1.make_choice("камень")
player2.make_choice("ножницы")


game = Game()
game.play(player1, player2)
```
