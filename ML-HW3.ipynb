```python
#DOI 10.24432/C5HP4Z

from ucimlrepo import fetch_ucirepo 
  
# fetch dataset 
breast_cancer_wisconsin_original = fetch_ucirepo(id=15) 
  
# data (as pandas dataframes) 
Xx = breast_cancer_wisconsin_original.data.features 
yy = breast_cancer_wisconsin_original.data.targets 
  
# metadata 
print(breast_cancer_wisconsin_original.metadata) 
  
# variable information 
print(breast_cancer_wisconsin_original.variables) 
```

```python
import pandas as pd
import numpy as np

bc = pd.concat([Xx, yy], axis=1)
bc.isnull().sum()
```

```python
bc.info()
```

```python
bc.head()
```

```python
median_bare_nuclei = bc['Bare_nuclei'].median()
bc['Bare_nuclei'].fillna(median_bare_nuclei, inplace=True)

bc.isnull().sum()
```

```python
bc.head()
```

```python
bc.drop_duplicates(inplace=True)
```

```python
#Метод IQR (межквартильный размах) — это способ определения выбросов, основанный на квартилях. 
#Он измеряет дисперсию точек данных между первой и третьей квартильными отметками.

# 1. Вычисление Q1 и Q3
Q1 = bc.quantile(0.25)
Q3 = bc.quantile(0.75)
IQR = Q3 - Q1

# Убираем выбросы
bc_filtered = bc[~((bc < (Q1 - 1.5 * IQR)) | (bc > (Q3 + 1.5 * IQR))).any(axis=1)]
bc_filtered
```
