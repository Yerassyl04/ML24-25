```python
from ucimlrepo import fetch_ucirepo
import pandas as pd

# Fetch dataset
heart_disease = fetch_ucirepo(id=45)

X = heart_disease.data.features
y = heart_disease.data.targets

df = pd.concat([X, y], axis=1)

df

# Print metadata and variable information
print(heart_disease.metadata)
print(heart_disease.variables)
```

```python
df.rename(columns={
    'age': 'age',
    'sex': 'sex',
    'cp': 'chest_pain',
    'trestbps': 'blood_pressure',
    'chol': 'cholesterol',
    'fbs': 'blood_sugar',
    'restecg': 'ecg_esults',
    'thalach': 'max_heart_rate',
    'exang': 'exercise-induced_angina',
    'oldpeak': 'st_depression',
    'slope': 'slope_peak_exercise',
    'ca': 'num_major_vessels',
    'thal': 'thalassemia',
}, inplace=True)
```

```python
df_clean = df.dropna(subset=['num_major_vessels', 'thalassemia'])
```

```python
df_clean = df_clean.copy()
```

```python
import pandas as pd

# Assuming df_clean is your DataFrame containing the dataset

# Blood pressure categorization (using integers instead of terms)
bins_blood_pressure = [0, 120, 130, 140, float('inf')]
labels_blood_pressure = [0, 1, 2, 3]  # Using integers instead of terms
df_clean['blood_pressure_cat'] = pd.cut(df_clean['blood_pressure'], bins=bins_blood_pressure, labels=labels_blood_pressure, right=False)

# Max heart rate categorization (using integers instead of terms)
bins_max_heart_rate = [0, 133, 153, 166, float('inf')]
labels_max_heart_rate = [0, 1, 2, 3]  # Using integers instead of terms
df_clean['max_heart_rate_cat'] = pd.cut(df_clean['max_heart_rate'], bins=bins_max_heart_rate, labels=labels_max_heart_rate, right=False)

# Now df_clean contains two new categorical columns: 'blood_pressure_cat' and 'max_heart_rate_cat'
# Drop the original columns 'blood_pressure' and 'max_heart_rate'
df_clean = df_clean.drop(columns=['blood_pressure', 'max_heart_rate'])

# Final df_clean will have the new categorical columns with integer labels

```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X = df_clean.drop(columns=['num'])  
y = df_clean['num'] 

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=45)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report


log_reg = LogisticRegression()
log_reg.fit(X_train_scaled, y_train)
log_reg_preds = log_reg.predict(X_test_scaled)

svm = SVC()
svm.fit(X_train_scaled, y_train)
svm_preds = svm.predict(X_test_scaled)

decision_tree = DecisionTreeClassifier()
decision_tree.fit(X_train_scaled, y_train)
decision_tree_preds = decision_tree.predict(X_test_scaled)


print("Logistic Regression Results:")
print(f"Accuracy: {accuracy_score(y_test, log_reg_preds):.4f}")
print(classification_report(y_test, log_reg_preds))

print("\nSupport Vector Machine Results:")
print(f"Accuracy: {accuracy_score(y_test, svm_preds):.4f}")
print(classification_report(y_test, svm_preds))

print("\nDecision Tree Results:")
print(f"Accuracy: {accuracy_score(y_test, decision_tree_preds):.4f}")
print(classification_report(y_test, decision_tree_preds))
```
