```python
# Determine the type of each column
def column_type_analysis(df):
    column_types = {}
    
    for col in df.columns:
        unique_values = df[col].nunique()
        dtype = df[col].dtype
        
        if dtype == 'object' or dtype.name == 'category':
            if unique_values == 2:
                column_types[col] = 'Binary Categorical'
            else:
                column_types[col] = 'Categorical'
        elif pd.api.types.is_numeric_dtype(df[col]):
            if unique_values == 2:
                column_types[col] = 'Binary Numeric'
            else:
                column_types[col] = 'Numeric'
        else:
            column_types[col] = 'Other'
    
    return column_types

# Get column types
column_types = column_type_analysis(df)

# Print results
for col, col_type in column_types.items():
    print(f"{col}: {col_type}")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X = df_clean.drop(columns=['num'])  
y = df_clean['num'] 

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=45)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report


log_reg = LogisticRegression()
log_reg.fit(X_train_scaled, y_train)
log_reg_preds = log_reg.predict(X_test_scaled)

svm = SVC()
svm.fit(X_train_scaled, y_train)
svm_preds = svm.predict(X_test_scaled)

decision_tree = DecisionTreeClassifier()
decision_tree.fit(X_train_scaled, y_train)
decision_tree_preds = decision_tree.predict(X_test_scaled)


print("Logistic Regression Results:")
print(f"Accuracy: {accuracy_score(y_test, log_reg_preds):.4f}")
print(classification_report(y_test, log_reg_preds))

print("\nSupport Vector Machine Results:")
print(f"Accuracy: {accuracy_score(y_test, svm_preds):.4f}")
print(classification_report(y_test, svm_preds))

print("\nDecision Tree Results:")
print(f"Accuracy: {accuracy_score(y_test, decision_tree_preds):.4f}")
print(classification_report(y_test, decision_tree_preds))
```

```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import SelectKBest, f_classif, mutual_info_classif
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score

# Assuming previous data preprocessing steps
# Rerun data preparation
X = df_clean.drop(columns=['num'])
y = df_clean['num']

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=45)

# Scale the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# 2. Mutual Information Feature Selection
selector_mi = SelectKBest(score_func=mutual_info_classif, k=10)
X_train_mi = selector_mi.fit_transform(X_train_scaled, y_train)
X_test_mi = selector_mi.transform(X_test_scaled)

# Get selected feature names
selected_features_mi = X_train.columns[selector_mi.get_support()].tolist()
print("\nTop 10 features (Mutual Information):", selected_features_mi)

# Train and evaluate model with selected features
lr_mi = LogisticRegression(max_iter=1000)
lr_mi.fit(X_train_mi, y_train)
y_pred_mi = lr_mi.predict(X_test_mi)
print("Accuracy with Mutual Information feature selection:", accuracy_score(y_test, y_pred_mi))
```
