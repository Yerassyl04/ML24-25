8. соз. массив с размером 5х5, запл. случ. числами от 1 до 100, и найдите позиции всех знач. превышающих 50

```python
i = np.random.randint(0, 100, (5, 5))
pos= np.argwhere(i > 50)

print(i)
print(pos)
```

9. соз. массив 7х7 и замените все элем, кратные 3, на число -1

```python
import numpy as np

j= np.random.randint(1, 101, (7, 7))

j[j % 3 == 0] = -1

print(j)
```

10. соз. массив и найдите суммы всех его четных элем.
